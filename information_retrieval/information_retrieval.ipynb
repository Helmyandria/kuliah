```python
articles = [
   'Pada hari minggu saya pergi ke pasar malam malam',
   'Malam itu di pasar ada diskon besar besaran',
   'Di pasar malam terdapat banyak permainan',
   'Malam minggu di pasar terjadi kebakaran',
   'Ada mainan terbuat dari malam'
]
print 'article : '
for article in articles:
    print article
# dapatkan semua kata dalam articles
words = []
for article in articles:
    article = article.split(' ')
    for word in article:
        if word not in words:
            words.append(word)
# stemming
stem = {'besaran' : 'besar', 
        'terbuat' : 'buat',
        'terdapat' : 'dapat',
        'kebakaran' : 'bakar',
        'mainan' : 'main',
        'permainan' : 'main',
        'terjadi' : 'jadi'}
attributes = ['di', 'itu', 'pada', 'ke', 'dari', 'pada']
new_words = []
for i, word in enumerate(words):
    if word in stem:
        word = stem[word]
    word = word.lower()
    if (word not in new_words) and (word not in attributes):
        new_words.append(word)
words = new_words
print '\njumlah kata'
print len(words)
print '\nkata'
print words

# vectorization
def vectorization(articles):
    vectors = []
    for article in articles:
        vector = []
        for i in range(len(words)):
            vector.append(0)
        article_words = article.split(' ')
        for article_word in article_words:
            if article_word in attributes:
                continue
            article_word = article_word.lower()
            for key in stem:
                if key == article_word:
                    article_word = stem[key]
            for i, word in enumerate(words):
                if word == article_word:
                    vector[i] += 1
        vectors.append(vector)
    return vectors


vectors = vectorization(articles)
print '\nvectors of articles : '
for vector in vectors:
    print vector
    
keyword = 'pasar malam'
vector_keyword = vectorization([keyword])[0]
print '\nkeyword '
print keyword
print '\nvector keyword'
print vector_keyword
```

Vector
======

$$ a=
\begin{bmatrix}
    a_{x} \\
    a_{y} \\
    a_{z} \\
\end{bmatrix}
$$

$$|a| = \sqrt{a_x^2 + b_x^2}$$

Dot product?
============
$$a \cdot b = |a| \times |b| \cos(\theta) $$

and

$$a \cdot b = (a_x \times b_x) + (a_y \times b_y) $$

Therefore
=========
$$\cos(\theta) = \frac{(a_x \times b_x) + (a_y \times b_y)}{|a| \times |b|}$$


```python
import math

def length(vector):
    cummulation = 0
    for number in vector:
        cummulation += number * number
    return cummulation ** 0.5

def dot_product(vector_1, vector_2):
    cummulation = 0
    for i, num_1 in enumerate(vector_1):
        num_2 = vector_2[i]
        cummulation += num_1 * num_2
    return cummulation

for i,vector in enumerate(vectors):
    print articles[i]
    print vector
    cos_theta = dot_product(vector, vector_keyword)/(length(vector)*length(vector_keyword))
    print cos_theta
    print math.acos(cos_theta)
```

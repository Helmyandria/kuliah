```python
articles = [
   'Pada hari minggu saya pergi ke pasar malam malam',
   'Malam itu di pasar ada diskon besar besaran',
   'Di pasar malam terdapat banyak permainan',
   'Malam minggu di pasar terjadi kebakaran',
   'Ada mainan terbuat dari malam'
]

unique_words = []
stem = {
    'besaran' : 'besar',
    'mainan'  : 'main',
    'permainan' : 'main',
    'kebakaran' : 'bakar',
    'terjadi'   : 'jadi',
    'terbuat'   : 'buat',
    'terdapat'  : 'dapat',
    'di'        : '',
    'dan'       : '',
    'ke'        : '',
    'dari'      : '',
    'pada'      : '',
    'itu'       : '',
}
for article in articles:
    words = article.split(' ')
    for word in words:
        word = word.lower()
        for key in stem:
            if word == key:
                word = stem[key]
        if word == '':
            continue
        if word not in unique_words:
            unique_words.append(word)
print len(unique_words)
print unique_words

def vectorize(articles):
    vectors = []
    for article in articles:
        vector = []
        for i in range(len(unique_words)):
            vector.append(0)
        words = article.split(' ')
        for word in words:
            word = word.lower()
            for key in stem:
                if word == key:
                    word = stem[key]
            if word == '':
                continue
            for i, unique_word in enumerate(unique_words):
                if unique_word == word:
                    vector[i] += 1
                    break
        vectors.append(vector)
    return vectors

vectors = vectorize(articles)
print ""
for i, article in enumerate(articles):
    print article
    print vectors[i]

keyword = 'Pasar Malam Minggu'
keyword_vector = vectorize([keyword])[0]
print ""
print keyword
print keyword_vector
```

Vector
======

$$ a=
\begin{bmatrix}
    a_{x} \\
    a_{y} \\
    a_{z} \\
\end{bmatrix}
$$

$$|a| = \sqrt{a_x^2 + b_x^2}$$

Dot product?
============
$$a \cdot b = |a| \times |b| \cos(\theta) $$

and

$$a \cdot b = (a_x \times b_x) + (a_y \times b_y) $$

Therefore
=========
$$\cos(\theta) = \frac{(a_x \times b_x) + (a_y \times b_y)}{|a| \times |b|}$$


```python
import math

def length(vector):
    cummulation = 0
    for number in vector:
        cummulation += number * number
    return cummulation ** 0.5

def dot_product(vector_1, vector_2):
    cummulation = 0
    for i, num_1 in enumerate(vector_1):
        num_2 = vector_2[i]
        cummulation += num_1 * num_2
    return cummulation

def cos_theta(vector_1, vector_2):
    return dot_product(vector_1, vector_2)/(length(vector_1) * length(vector_2))

def theta_value(vector_1, vector_2):
    return math.acos(cos_theta(vector_1, vector_2))

for i, article in enumerate(articles):
    vector = vectors[i]
    theta = theta_value(vector, keyword_vector)
    print article
    print vector
    print theta
    print ''
```

```python
def get_unique_words(articles):
    unique_words = []
    stem = {
        'besaran' : 'besar',
        'mainan'  : 'main',
        'permainan' : 'main',
        'kebakaran' : 'bakar',
        'terjadi'   : 'jadi',
        'terbuat'   : 'buat',
        'terdapat'  : 'dapat',
        'di'        : '',
        'dan'       : '',
        'ke'        : '',
        'dari'      : '',
        'pada'      : '',
        'itu'       : '',
    }
    for article in articles:
        words = article.split(' ')
        for word in words:
            word = word.lower()
            for key in stem:
                if word == key:
                    word = stem[key]
            if word == '':
                continue
            if word not in unique_words:
                unique_words.append(word)
    return unique_words

def AND(array1, array2):
    result = []
    for i in range(len(array1)):
        result.append(array1[i] and array2[i])
    return result

def OR(array1, array2):
    result = []
    for i in range(len(array1)):
        result.append(array1[i] or array2[i])
    return result

def NOT(array):
    result = []
    for i in range(len(array)):
        result.append(not array[i])
    return result

articles = [
   'Pada hari minggu saya pergi ke pasar malam malam',
   'Malam itu di pasar ada diskon besar besaran',
   'Di pasar malam terdapat banyak permainan',
   'Malam minggu di pasar terjadi kebakaran',
   'Ada mainan terbuat dari malam'
]
unique_words = get_unique_words(articles)
print "Kata-kata unik: "
print unique_words
print ""
print "Jumlah kata-kata unik: " + str(len(unique_words))
```

```python
stem = {
    'besaran' : 'besar',
    'mainan'  : 'main',
    'permainan' : 'main',
    'kebakaran' : 'bakar',
    'terjadi'   : 'jadi',
    'terbuat'   : 'buat',
    'terdapat'  : 'dapat',
    'di'        : '',
    'dan'       : '',
    'ke'        : '',
    'dari'      : '',
    'pada'      : '',
    'itu'       : '',
}
boolean_indexes = []
for unique_word in unique_words:
    boolean_index = []
    for article in articles:
        found = False
        words = article.split(' ')
        for word in words:
            word = word.lower()
            for key in stem:
                if word == key:
                    word = stem[key]
            if word == '':
                continue
            if word == unique_word:
                found = True
                break
        boolean_index.append(found)
    boolean_indexes.append(boolean_index)
    
for article in articles:
    print article
print "\nBOOLEAN INDEX:"
for i, unique_word in enumerate(unique_words):
    print unique_word + " " + str(boolean_indexes[i])
```

```python
# (malam AND NOT diskon) OR bakar

def find_boolean_index(word):
    for i, unique_word in enumerate(unique_words):
        if word == unique_word:
            return boolean_indexes[i]

malam = find_boolean_index('malam')
diskon = find_boolean_index('diskon')
bakar = find_boolean_index('bakar')
minggu = find_boolean_index('minggu')
pasar = find_boolean_index('pasar')

result = OR(bakar, AND(pasar, NOT(minggu)) )
print "Result : "
print result
for i, article in enumerate(articles):
    if result[i]:
        print article
```

## Porter Algorithm
Rule
* `SSES -> SS`, 

    E.g: caresses -> caress
    
    
* `IES -> I`, 

    E.g: ponies -> poni
    
    
* `SS --> SS`, 

    E.g: caress -> caress
    
    
* `S --> (ommited)`, 

    E.g: cats -> cat

```python
# Quicksort

def quicksort(angka):
    pivot = angka[0]
    left, right = [], []
    for i in range(1,len(angka)):
        if angka[i]<pivot:
            left.append(angka[i])
        else:
            right.append(angka[i])
    if len(left) > 1:
        left = quicksort(left)
    if len(right) > 1:
        right = quicksort(right)
    print left, pivot, right
    return left + [pivot] + right

a = [5,4,3,2,1,9,1,2,4,5,6,7,8,1,3]
print "angka : ", a
print "Hasil : ", quicksort(a)
```

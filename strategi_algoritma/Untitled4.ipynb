```python
NODES = ['A','B','C','D','E','F']
EDGES = {
    'A' : {'B' : 4, 'C' : 3, 'E' : 9},
    'B' : {'A' : 4, 'C' : 3, 'D' : 3},
    'C' : {'A' : 3, 'B' : 3, 'D' : 3, 'E' : 4},
    'D' : {'B' : 3, 'C' : 3, 'E' : 2},
    'E' : {'A' : 9, 'C' : 4, 'D' : 2, 'F' : 2},
    'F' : {}
}
START = 'A'
END = 'F'
PATH = [[START]]

def get_distance(path):
    total = 0
    for i in range(len(path)-1):
        total += EDGES[path[i]][path[i+1]]
    return total

def get_path(path_list):
    new_path_list = []
    for path in path_list:
        found = False
        last_node = path[-1]
        for next_node in EDGES[last_node]:
            if next_node not in path:
                new_path = list(path)
                new_path.append(next_node)
                new_path_list.append(new_path)
                found = True
        if not found:
            new_path_list.append(path)
    if new_path_list == path_list:
        return [x for x in path_list if x[-1] == END]
    return get_path(new_path_list)

print 'POSSIBLE SOLUTION : '
PATH = get_path(PATH)
for path in PATH:
    print path

# temukan jarak terpendek
min_distance = -1
for path in PATH:
    if get_distance(path) < min_distance or min_distance == -1:
        min_distance = get_distance(path)
# filter path, hanya yang jaraknya samadengan jarak terpendek
PATH = [x for x in PATH if get_distance(x) == min_distance]
print 'BEST SOLUTION : '
for path in PATH:
    print path, get_distance(path)
    
```

```python
PETA = [
    [1,1,1,0,0,0,5],
    [1,0,1,0,0,0,0],
    [1,1,1,2,0,0,1],
    [1,1,1,1,1,1,2]
]
def get_coverage(r1, c1, r2, c2):
    memo = {}
    for i in range(r1-1,r1+2):
        for j in range(c1-1, c1+2):
            if i<0 or i>=len(PETA) or j<0 or j>=len(PETA[0]):
                continue
            memo[(i,j)] = PETA[i][j]
    for i in range(r2-1,r2+2):
        for j in range(c2-1, c2+2):
            if i<0 or i>=len(PETA) or j<0 or j>=len(PETA[0]):
                continue
            memo[(i,j)] = PETA[i][j]
    total = 0
    for key in memo:
        total += memo[key]
    return total

SOLUTION = {}
best_coverage = -1
for r1 in range(0,len(PETA)):
    for c1 in range(0,len(PETA[0])):
        for r2 in range(0,len(PETA)):
            for c2 in range(0,len(PETA[0])):
                coverage = get_coverage(r1,c1,r2,c2)
                SOLUTION[(r1,c1,r2,c2)] = coverage
                if coverage > best_coverage or best_coverage == -1:
                    best_coverage = coverage

print 'FOUND SOLUTIONS : ', len(SOLUTION)
for x in SOLUTION:
    print x, ', coverage : ', SOLUTION[x]

print 'FOUND BEST SOLUTION : '
BEST_SOLUTION = [x for x in SOLUTION if SOLUTION[x] == best_coverage]
for x in BEST_SOLUTION:
    print x, SOLUTION[x]
```

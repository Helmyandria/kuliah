```python
NODES = ["A", "B", "C", "D", "E", "F", "G"]
EDGES = {
    "A" : {"B":5, "C":4, "D":9},
    "B" : {"A":5, "C":3, "E":2},
    "C" : {"A":4, "B":3, "D":3},
    "D" : {"A":9, "C":3, "F":1},
    "E" : {"B":2, "F":2, "G":1},
    "F" : {"D":1, "E":2},
    "G" : {"E":1}
}
START = "A"
END = "F"
PATH = [[START]]

def get_distance(path, edges):
    total = 0
    for i in range(len(path)-1):
        total += edges[path[i]][path[i+1]]
    return total

def get_path(path_list, edges, end):
    new_path_list = []
    for path in path_list:
        last_node = path[-1]
        found = False
        if last_node != end:
            for node in edges[last_node]:
                if node not in path:
                    found = True
                    new_path = list(path)
                    new_path.append(node)
                    new_path_list.append(new_path)
        if not found:
            new_path_list.append(path)
    if new_path_list == path_list:
        return [x for x in new_path_list if x[-1]== end]
    return get_path(new_path_list,edges, end)

valid_path =  get_path([[START]], EDGES, END)
min_distance = -1
for path in valid_path:
    if get_distance(path, EDGES) < min_distance or min_distance == -1:
        min_distance = get_distance(path, EDGES)

print min_distance
for path in valid_path:
    if min_distance == get_distance(path, EDGES):
        print path
```

```python
PETA = [
    [4,0,2,1,0,1,1],
    [0,0,1,1,0,3,2],
    [0,2,0,1,2,0,0],
    [1,0,4,2,1,2,0],
    [3,0,0,0,0,3,4],
    [0,0,0,2,0,1,1]
]

def get_coverage(r1,c1,r2,c2, peta):
    tmp, total = {}, 0
    for r in range(r1-1,r1+2):
        if r>=0 and r<len(peta):
            for c in range(c1-1, c1+2):
                if c>=0 and c<len(peta[0]):
                    tmp[(r,c)] = peta[r][c]
    for r in range(r2-1,r2+2):
        if r>=0 and r<len(peta):
            for c in range(c2-1, c2+2):
                if c>=0 and c<len(peta[0]):
                    tmp[(r,c)] = peta[r][c]
    for x in tmp:
        total += tmp[x]
    return total

solutions = []
for r1 in range(len(PETA)):
    for c1 in range(len(PETA[0])):
        for r2 in range(r1,len(PETA)):
            for c2 in range(c1,len(PETA[0])):
                solutions.append((r1,c1,r2,c2))

best_solutions = []
best_coverage = -1
print "Solution found: ", len(solutions)
for solution in solutions:
    r1,c1,r2,c2 = solution
    coverage = get_coverage(r1,c1,r2,c2, PETA)
    print solution, coverage
    if coverage > best_coverage:
        best_coverage = coverage

for solution in solutions:
    r1,c1,r2,c2 = solution
    coverage = get_coverage(r1,c1,r2,c2, PETA)
    if coverage == best_coverage:
        best_solutions.append(solution)
        
print "Best: "
for solution in best_solutions:
    print solution, best_coverage
```
